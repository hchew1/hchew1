### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
# Display column data types
print("Column Data Types:")
print(df.dtypes)

# Display counts of non-missing values for each column
print("\nCounts of Non-Missing Values:")
print(df.count())

# Display counts of missing values for each column
print("\nCounts of Missing Values:")
print(df.isnull().sum())
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
# Generate basic statistics for numerical columns
numerical_stats = df.describe()
display(numerical_stats)
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
# Display unique values and their counts in the Country column
country_counts = df['Country'].value_counts()
display(country_counts)
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# Standardize entries in the Country column
country_mapping = {
    'United States of America': 'USA',
    'United States': 'USA',
    'UK': 'United Kingdom',
    'England': 'United Kingdom',
    # Add more mappings as needed
}
df['Country'] = df['Country'].replace(country_mapping)

# Standardize entries in the EdLevel column
edlevel_mapping = {
    'Bachelor’s degree': 'Bachelor',
    'Master’s degree': 'Master',
    'Some college/university study without earning a degree': 'Some College',
    # Add more mappings as needed
}
df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

# Display the DataFrame after standardization
display(df)
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
# Encode the Employment column using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Employment'])

# Display the DataFrame after encoding
display(df_encoded)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here

# Calculate the number of missing values for each column
missing_values = df.isnull().sum()

# Identify columns with the highest number of missing values
columns_with_most_missing = missing_values[missing_values > 0].sort_values(ascending=False)

# Display the columns with the highest number of missing values
print("Columns with the highest number of missing values:")
print(columns_with_most_missing)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
# Impute missing values in the ConvertedCompYearly column with the mean
mean_converted_comp = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly'].fillna(mean_converted_comp, inplace=True)

# Alternatively, impute missing values with the median
# median_converted_comp = df['ConvertedCompYearly'].median()
# df['ConvertedCompYearly'].fillna(median_converted_comp, inplace=True)

# Display the DataFrame after imputation
display(df)
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# Impute missing values in the RemoteWork column with the most frequent value
most_frequent_remote_work = df['RemoteWork'].mode()[0]
df['RemoteWork'].fillna(most_frequent_remote_work, inplace=True)

# Display the DataFrame after imputation
display(df)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
# Apply Min-Max normalization to the ConvertedCompYearly column
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())

# Display the DataFrame after normalization
display(df)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np
```

```python
## Write your code here
# Apply log transformation to the ConvertedCompYearly column
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)  # Adding 1 to avoid log(0)

# Display the DataFrame after log transformation
display(df)
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
# Define a function to convert age ranges to numeric midpoints
def convert_age_range(age):
    if pd.isnull(age):
        return None
    if '-' in age:
        start, end = age.split('-')
        return (int(start) + int(end)) / 2
    elif 'older' in age:
        return 65  # Assuming '65 years or older' as 65
    elif 'younger' in age or 'Under' in age:
        return 18  # Assuming 'Under 18 years old' as 18
    elif 'years old' in age:
        return int(age.split()[0])
    else:
        return int(age)

# Apply the function to the Age column
df['Age'] = df['Age'].apply(convert_age_range)

# Calculate the mean of the Age column
mean_age = df['Age'].mean()
print(f"The mean of the Age column is: {mean_age}")

# # Define a function to categorize experience levels
# def categorize_experience(years):
#     if pd.isnull(years):
#         return None
#     years = float(years)
#     if years < 1:
#         return 'Beginner'
#     elif 1 <= years < 5:
#         return 'Intermediate'
#     elif 5 <= years < 10:
#         return 'Advanced'
#     else:
#         return 'Expert'

# # Apply the function to the YearsCodePro column to create ExperienceLevel
# df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)

# Display the DataFrame after adding the ExperienceLevel column
display(df)
```
